```python
# Store target variable
survived_train = df_train.Survived
```

```python
# Concatenate train and test sets to preprocess and clean data
data = pd.concat([df_train.drop(['Survived'], axis=1), df_test])
```

```python
# Describe the structure of the concatenated Dataframe
data.info()
```

```python
# Fill missing values on Age and Fare using their median values
data['Age'] = data.Age.fillna(data.Age.median())
data['Fare'] = data.Fare.fillna(data.Fare.median())
```

```python
# Describe the strucutre with filled values
data.info()
```

```python
# Get dummy variables for features that will be used to train models
data = pd.get_dummies(data, columns=['Sex'], drop_first=True)
```

```python
# Check the contents of the Dataframe
data.head()
```

```python
# Select features that will be used on machine learning models
data = data[['Sex_male', 'Fare', 'Age', 'Pclass', 'SibSp']]
```

```python
# Check the contents of the Dataframe
data.head()
```

```python
# Describe the strucutre of the Dataframe
data.info()
```

```python
# Split the treated data into training and test sets
data_train = data[:891]
data_test = data[891:]
```

```python
# Get numpy arrays from train and test sets to be used with scikit-learn models
X = data_train.values
test = data_test.values
y = survived_train.values
```

```python
# Create a decision tree classifier
clf = tree.DecisionTreeClassifier(max_depth=3)
```

```python
# Fit the data into the model
clf.fit(X, y)
```

```python
# Predict on the test set
y_pred = clf.predict(test)
```

```python
# Store the predictions on the test dataframe
df_test['Survived'] = y_pred
```

```python
# Explore different values for the hyperparameter max_depth. Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)
```

```python
# Create arrays to store model accuracies
dep = np.arange(1, 9)
train_accuracy = np.empty(len(dep))
test_accuracy = np.empty(len(dep))
```

```python
# Register the accuracies
for i, k in enumerate(dep):
    clf = tree.DecisionTreeClassifier(max_depth=k)
    clf.fit(X_train, y_train)
    train_accuracy[i] = clf.score(X_train, y_train)
    test_accuracy[i] = clf.score(X_test, y_test)
```

```python
# Show the accuracies in a plot
plt.title('clf: varying depth of tree')
plt.plot(dep, test_accuracy, label='Testing accuracy')
plt.plot(dep, train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Depth of tree')
plt.ylabel('Accuracy')
plt.show()
```
