# Tipos de gráficos

Exemplos de gráficos montados no matplotlib.

## Importação das bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Preparação do dataset

```python
# Base de dados
df_data = {
    'Nome': ['João', 'Maria', 'Guilherme', 'Samantha', 'Fernando', 'Luísa', 'Alexandre', 'Felipe', 'Isadora', 'Jaqueline'],
    'Idade': [18, 19, 19, 18, 20, 19, 18, 18, 18, 21],
    'Sexo': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'M', 'F', 'F'],
    'Q1 Nota': [6, 7, 10, 7, 9, 6, 4, 4, 7, 9],
    'Q2 Nota': [3, 6, 6, 7, 8, 7, 4, 3, 9, 10]
}
```

```python
df = pd.DataFrame(df_data)
```

```python
df.head()
```

```python
df = df.melt(id_vars=['Nome', 'Idade', 'Sexo'], value_vars=['Q1 Nota', 'Q2 Nota'], var_name='Questão', value_name='Nota')
```

```python
df['Questão'] = df['Questão'].str.extract(r'(^Q\d+)')
```

```python
df.head()
```

## Visualização dos dados

```python
plt.figure(figsize=(15, 5))
plt.scatter(df.loc[df['Questão'] == 'Q1']['Nota'], df.loc[df['Questão'] == 'Q2']['Nota'])
plt.title('Relacionamento entre as questões 1 e 2')
plt.xlabel('Questão 1 (Q1)')
plt.ylabel('Questão 2 (Q2)')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
ax = plt.axes()
plt.boxplot([df.loc[df['Sexo'] == 'M',:]['Nota'], df.loc[df['Sexo'] == 'F',:]['Nota']])
plt.title('Distribuição das notas')
plt.xlabel('Sexo')
ax.set_xticklabels(['Masculino', 'Feminino'])
plt.ylabel('Nota')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.hist(df['Nota'])
plt.title('Distribuição das notas')
plt.xlabel('Nota')
plt.ylabel('Quantidade de alunos')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.pie(df['Sexo'].value_counts(), labels=['Masculino', 'Feminino'], normalize=True)
plt.legend(loc='lower right')
plt.title('Proporção de sexos')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.bar(df['Idade'].unique(), df['Idade'].value_counts() / len(df['Questão'].unique()))
plt.title('Distribuição de idades')
plt.xlabel('Idade')
plt.ylabel('Quantidade de alunos')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.heatmap(df.corr(), annot=True)
plt.title('Correlação entre a idade e a nota')
plt.show()
```
