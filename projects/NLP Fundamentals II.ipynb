```python
# Define functions to automate word count and proportion calculation tasks
def lexical_diversity(text):
    return len(set(text)) / len(text)

def percentage(count, total):
    return 100 * count / total

def average_words(text):
    return len(text) / len(set(text))
```

```python
# Test the functions
print('Lexical diversity (Moby Dick): {0}'.format(lexical_diversity(text1)))
print('Lexical diversity (Sense and Sensibility): {0}'.format(lexical_diversity(text2)))

print('Average words (Moby Dick): {0}'.format(average_words(text1)))
print('Percentage (Moby Dick): {0}'.format(percentage(text1.count('monstrous'), len(text1))))
```

```python
# Display statistics for each book available in the Gutenberg project
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(fileid, 'AVG word length: {0}.'.format(round(num_chars / num_words)), 'AVG sentence length: {0}'.format(round(num_words / num_sents)), 'AVG word usage: {0}'.format(round(num_words / num_vocab)))
```

```python
# Describe the genres available in the Brown corpus
brown.categories()
```

```python
# Inspect the occurencies of given words in the Brown corpus
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print(m + ':', fdist[m], end=' ')
```

```python
# Inspect the stylistics of the genres available in the Brown corpus

'''
The counts for each modal are displayed by genre in a table.
'''
cfd = nltk.ConditionalFreqDist((genre, word) for genre in brown.categories() for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```
