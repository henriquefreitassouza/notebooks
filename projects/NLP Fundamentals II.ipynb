```python
# Calculate the lexical richeness and the average word usage frequency of the Book of Genesis

'''Around 6% of the text is comprised of unique words and each word is used 16 times an average. The following
are two ways to get to the same conclusion.'''
print(len(set(text3)) / len(text3))
print(len(text3) / len(set(text3)))
```

```python
# Count the number of times a given word is used throughout the text
text3.count('smote')
```

```python
# Calculate the proportion of the text that is composed of a single word
text1.count('a') / len(text1)
```

```python
# Define functions to automate word count and proportion calculation tasks
def lexical_diversity(text):
    return len(set(text)) / len(text)

def percentage(count, total):
    return 100 * count / total

def average_words(text):
    return len(text) / len(set(text))
```

```python
# Test the functions
print('Lexical diversity (Moby Dick): {0}'.format(lexical_diversity(text1)))
print('Lexical diversity (Sense and Sensibility): {0}'.format(lexical_diversity(text2)))

print('Average words (Moby Dick): {0}'.format(average_words(text1)))
print('Percentage (Moby Dick): {0}'.format(percentage(text1.count('monstrous'), len(text1))))
```
