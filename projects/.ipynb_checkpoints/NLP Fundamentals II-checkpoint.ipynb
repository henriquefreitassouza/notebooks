```python
# Look for similar words based in the context for monstrous in Moby Dick and Sense and Sensibility

'''
In the second text, monstrous is a word used together with good adjectives and words, somehow intensifying the
meaning of these other words
'''
text1.similar('monstrous')
print('\n')
text2.similar('monstrous')
```

```python
# List common contexts between two words
text2.common_contexts(['monstrous', 'very'])
print('\n')
text1.common_contexts(['monstrous', 'fearless'])
```

```python
# Count the number of words (punctuation included) in the Book of Genesis

'''
In NLP, each word or punctuation is called a token, so there are 44764 tokens in the Book of Genesis. A token is
a sequence of characters.
'''
len(text3)
```

```python
# Sort the unique tokens in the Book of Genesis

'''Punctuation and capitalized words precede lowercase words.'''
sorted(set(text3))
```

```python
# Count the unique tokens in the Book of Genesis

'''This book has 2789 word types, which are unique ocurrencies of sentences in a text. Punctuation symbols are
called types.'''
len(set(text3))
```

```python
# Calculate the lexical richeness and the average word usage frequency of the Book of Genesis

'''Around 6% of the text is comprised of unique words and each word is used 16 times an average. The following
are two ways to get to the same conclusion.'''
print(len(set(text3)) / len(text3))
print(len(text3) / len(set(text3)))
```

```python
# Count the number of times a given word is used throughout the text
text3.count('smote')
```

```python
# Calculate the proportion of the text that is composed of a single word
text1.count('a') / len(text1)
```

```python
# Define functions to automate word count and proportion calculation tasks
def lexical_diversity(text):
    return len(set(text)) / len(text)

def percentage(count, total):
    return 100 * count / total

def average_words(text):
    return len(text) / len(set(text))
```

```python
# Test the functions
print('Lexical diversity (Moby Dick): {0}'.format(lexical_diversity(text1)))
print('Lexical diversity (Sense and Sensibility): {0}'.format(lexical_diversity(text2)))

print('Average words (Moby Dick): {0}'.format(average_words(text1)))
print('Percentage (Moby Dick): {0}'.format(percentage(text1.count('monstrous'), len(text1))))
```
