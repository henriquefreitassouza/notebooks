# Pyplot and Seaborn

Pyplot is a library similar to MATLAB. This notebook will describe some of its attributes and methods, as well as how to enhance its views with Seaborn.

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data manipulation

```python
# Plot an list. The items in the list represent the y axis and its indexes are plotted on the x axis
plt.ylabel('Some numbers')
plt.plot([1,2,3,4])
plt.show()
```

```python
# Build a graph with adjusted axis. The string ro means "red circle"
plt.ylabel('Some other numbers')
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,5,0,17])
plt.show()
```

```python
# Create multiple plots with a single command. r-- means "red dash", bs means "blue square" and g^ means "green triangle"
dataset = np.arange(0., 5., 0.2)
plt.plot(dataset, dataset, 'r--', dataset, dataset ** 2, 'bs', dataset, dataset ** 3, 'g^')
plt.show()
```

```python
# Create a new graph
x = np.arange(0,10)
y = np.arange(0,10)
plt.plot(x, y, linewidth=4.0)
plt.show()
```

```python
# Create a graph with two plots. bo means "blue circle"
dataset1 = y1 = np.arange(0, 1, 0.1)
dataset2 = y2 = np.arange(0, 1, 0.2)

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(dataset1, y1, 'bo')

plt.subplot(2, 1, 2)
plt.plot(dataset2, y2, 'r--')

plt.show()
```

```python
# Show graph with multiple figures, subplots and plots

plt.figure(1)
plt.subplot(2, 1, 1)
plt.plot([1, 2, 3])
plt.subplot(2, 1, 2)
plt.plot([4, 5, 2])

plt.figure(2)
plt.subplot(2, 1, 2)
plt.plot([6, 7, 8])

plt.title('Multiple plots')
plt.show()
```

```python
# Set seed for random functions
np.random.seed(42)
```

```python
# Mean and standard deviation
mu, sigma = 100, 15
```

```python
# Data points
x = mu + sigma * np.random.randn(10000)
```

```python
# Plot a histogram normally distributed. Mathematical terms are written using TEX expressions
n, bins, patches = plt.hist(x, bins=50, normed=1, facecolor='k', alpha=0.9)

plt.xlabel('Data points')
plt.ylabel('Probability')
plt.title('Histogram')
plt.text(45, .027, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
# Create a graph with annotations
plt.plot(np.arange(0,10), np.arange(0,10))
plt.annotate('Middle', xy=(4.5,4.5), xytext=(4,6.5), arrowprops=dict(facecolor='blue', shrink=0.05))
```

```python
# Create a histogram with specified scales
plt.hist(np.random.randn(5000), bins=20, normed=1)
plt.xlim(-5,5)
plt.ylim(0,0.5)
plt.show()
```

## Data manipulation with Seaborn

```python
# Import iris dataset
iris = sns.load_dataset('iris')
```

```python
# Inspect the loaded dataset
type(iris)
```

```python
# Check dataset contents
iris.head()
```

```python
# Visualize iris dataset in a swarmplot
sns.swarmplot(x='species', y='petal_length', data=iris)
plt.show()
```

```python
# Visualize iris dataset in a factorplot
sns.factorplot('species', 'petal_length', data=iris)
plt.show()
```

```python
# Check the available styles to use in graphics
plt.style.available
```

```python
# Change default context
sns.set_context('poster')
sns.swarmplot(x='species', y='petal_length', data=iris)
plt.show()
```

```python
# Print a box and whiskers plot
sns.boxplot(x='petal_length', data=iris)
plt.show()
```

```python
# Print a FacetGrid scatter plot
sns.lmplot('petal_width', 'petal_length', iris)
plt.show()
```

```python
# Reset seaborn styles
sns.set()
```

```python
# Print a correlation heatmap
sns.heatmap(np.corrcoef(np.random.rand(10,12)), cmap='YlGnBu', square=True)
plt.show()
```
