# Pandas fundamentals

This notebook demonstrates some of the features available in Pandas

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating a data structure

```python
# Create an array with 100 data points distributed along 20 rows and 5 columns
dataset = np.random.randint(0, 100, size=(20, 5))
```

## Loading the Dataframe

```python
# Create a dataframe with a numpy array with 100 entries
df = pd.DataFrame(dataset, columns=['Random 1', 'Random 2', 'Random 3', 'Random 4', 'Random 5'])
```

## Manipulation

```python
# Check the first 5 entries
df.head()
```

```python
# Describe statistical properties
df.describe()
```

```python
# Describe the datatype of each column and the total number of entries
df.info()
```

```python
# Create a range of dates to use as index on the Dataframe
dates = pd.date_range('20180101', periods=20)
```

```python
# Set a unique index for each line based on a date
df.index = dates
```

```python
# Check the last 5 entries on the Dataframe
df.tail()
```

```python
# Show the names of the columns
df.columns
```

```python
# Show the indexes
df.index
```

```python
# Show the values stored in the Dataframe
df.values
```

```python
# Sort by index descending
df.sort_index(axis=0, ascending=False).head()
```

```python
# Sort by value descending
df.sort_values(by='Random 2', ascending=False).head()
```

```python
# Extract a Series from the Dataframe
df['Random 1']
```

```python
# Select the first 10 days
df['20180101':'20180110']
```

```python
# Select all data on index 0 (the first date)
df.loc['20180101']
```

```python
# Select all data on the first 2 columns
df.loc[:,['Random 1', 'Random 2']].head()
```

```python
# Select the first line by position
df.iloc[0]
```

```python
# Select the first 5 rows and third column
df.iloc[:5,2:3]
```

```python
# Select values based on boolean condition
df[df['Random 1'] > 50]
```

```python
# Create a new Series object to add a column on the Dataframe. Indexes must match
new_column = pd.Series(np.arange(20), index=pd.date_range('20180101', periods=20))
```

```python
# Add new column to Dataframe
df['Arange'] = new_column
```

```python
# Check the new column
df.head()
```

```python
# Create empty list
date_names = []
```

```python
# Populate the list with dates
for i, x in enumerate(range(1,21)):
    date_names.append('January ' + str(x))
```

```python
# Create a Series object
new_column = pd.Series(date_names, index=pd.date_range('20180101', periods=20))
```

```python
# Add a column to the Dataframe
df['Date'] = new_column
```

```python
# Check the new added column
df.head()
```

```python
# Apply custom function along the columns
df.iloc[:,:5].apply(np.sum)
```

```python
# Fill some values with nan
df.iloc[[2, 6, 7], [2, 3]] = np.NaN
```

```python
# Drop any rows with empty data
df.dropna(how='any')
```

```python
# Fill empty data with 10
df.fillna(value=10)
```

```python
# Mean value for each column
df.mean()
```

```python
# Mean value of rows
df.mean(1)
```

```python
# Shift the first 10 rows
df.shift(10)
```

```python
# Count the number of ocurrencies at the specified column
df['Random 1'].value_counts()
```

```python
# Change the string column to uppercase
df['Date'].str.upper()
```

```python
# Create a list of dataframes based on the previous one
df_temp = [df[:5], df[7:10]]
```

```python
# Check the list of Dataframes
df_temp
```

```python
# Concatenate the list of Dataframes
pd.concat(df_temp)
```

```python
# Create a Dataframe
left = pd.DataFrame({'date': pd.date_range('20180101', periods=2), 'left': [912.88, 444.11]})
```

```python
# Check Dataframe's contents
left.head()
```

```python
# Create a Dataframe to be joined with the previous one
right = pd.DataFrame({'date': pd.date_range('20180101', periods=2), 'right': [101.15, 662.38]})
```

```python
# Check Dataframe's contents
right.head()
```

```python
# Join two Dataframes
pd.merge(left, right, on='date')
```

```python
# Create a new Dataframe
df_1 = pd.DataFrame(np.random.randint(0, 20, size=(10, 2)), columns=['Random 1', 'Random 2'])
```

```python
# Check Dataframe's contents
df_1.head()
```

```python
# Create a new Dataframe to be appended to the previous one
df_2 = pd.DataFrame(np.random.randint(0, 10, size=(5, 2)), columns=['Random 1', 'Random 2'])
```

```python
# Check Dataframe's contents
df_2.head()
```

```python
# Append the second Dataframe to the first. Ignore index will fix duplicate indexes
df_1.append(df_2, ignore_index=True)
```

```python
# Group by column
df.groupby('Random 2').sum()
```

```python
# Group by multiple columns
df.groupby(['Random 3', 'Random 4']).sum()
```

```python
# Stack a Dataframe. The result is a Series object
df.stack()
```

```python
# Unstack a Dataframe.
df.unstack()
```

```python
# Create a pivot table based on a DataFrame
pd.pivot_table(df, values=['Random 1', 'Random 2', 'Random 3', 'Random 4', 'Random 5'], columns='Date').head()
```

```python
# Create a new figure object
plt.figure()
```

```python
# Plot the figure
df.plot()
```

```python
# Export data to csv
df.to_csv('pandas_data.csv')
```
