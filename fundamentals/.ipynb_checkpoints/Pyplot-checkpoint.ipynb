# Pyplot

Pyplot is a library similar to MATLAB. This notebook will describe some of its attributes and methods.

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data manipulation

```python
# Plot an list
plt.ylabel('Some numbers')
plt.plot([1,2,3,4])
plt.show()
```

```python
# Build a graph with adjusted axis. The string ro means "red circle"
plt.ylabel('Some other numbers')
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20])
plt.show()
```
