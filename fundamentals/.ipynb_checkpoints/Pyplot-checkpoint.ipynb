# Pyplot fundamentals

Pyplot is Matplotlib's library for plotting data in a variety of formats.

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
# Set inline plotting
%matplotlib inline
```

## Plotting data

```python
# Generate data and plot
t = np.arange(0.0, 0.2, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)', title='First plot')
ax.grid()

plt.show()
```

```python
# Working with subplots
x1 = np.linspace(0.0, 0.5)
x2 = np.linspace(0.0, 0.2)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('Two plots')
plt.ylabel('Oscilation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
# Customizing histograms

np.random.seed(19680801)

mu = 100
sigma = 15
x = mu + sigma + np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots()

n, bins, patches = ax.hist(x, num_bins, normed=1)

y = mlab.normpdf(bins, mu, sigma)

ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title('Histogram of IQ: $\mu=100$, $\sigma=15$')

fig.tight_layout()
plt.show()
```
